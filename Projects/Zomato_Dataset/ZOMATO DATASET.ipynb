```python
##observation
#when rating is b/w 4.5 to 4.9----->Excellent
#when rating is b/w 4.0 to 4.4------->Very Good
#when rating is b/w 3.5 to 3.9------->Good
#when rating is b/w 3.0 to 3.4------->Average
#when rating is b/w 2.5 to 2.9------->Average
#when rating is b/w 2.0 to 2.4------->Poor
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
           
```
