```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,ConfusionMatrixDisplay
```

```python
X=df[["Age","Sex","BP","Cholesterol","Na_to_K"]]
y=df["Drug"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
dtree=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=42)
dtree
```

```python
dtree.fit(X_train,y_train)
```

```python
y_pred=dtree.predict(X_test)
y_pred
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n",confusion_matrix(y_test,y_pred)) 
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
conf_matrix=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=df["Drug"].unique())
conf_matrix
conf_matrix.plot(cmap="Blues", values_format='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
tree.plot_tree(dtree, filled=True, feature_names=["Age","Sex","BP","Cholesterol","Na_to_K"], 
               class_names=df["Drug"].unique())
plt.title("Decision Tree Visualization")
plt.show()
```
