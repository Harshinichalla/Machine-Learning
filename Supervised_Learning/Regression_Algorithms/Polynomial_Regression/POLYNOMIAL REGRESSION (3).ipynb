```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("C:\\Users\\chars\\Downloads\\archive (1)\\Ice_cream selling data.csv")
df.head()
```

```python
x=df["Temperature (°C)"].values.reshape(-1,1)    #to_frame()
y=df["Ice Cream Sales (units)"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
poly
```

```python
x_train_poly=poly.fit_transform(x_train)
x_train_poly
```

```python
x_test_poly=poly.transform(x_test)
x_test_poly
```

```python
model_poly=LinearRegression()
model_poly.fit(x_train_poly,y_train)
y_pred_poly=model_poly.predict(x_test_poly)
y_pred_poly
```

```python
plt.scatter(x,y,color='blue',label='Data')
plt.plot(x,model.predict(x),color='green',label='Linear Fit')
plt.plot(x,model_poly.predict(poly.transform(x)),color='red',label='Polynomial Fit')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```
