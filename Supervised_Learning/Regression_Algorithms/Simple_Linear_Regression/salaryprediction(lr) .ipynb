```python
import pandas as pd
```

```python
df=pd.read_csv("Salary Data.csv")
df
```

```python
df.isnull()
```

```python
df.head(20)
```

```python
df.isnull().sum() 
```

```python
df["Age"].fillna(0, inplace=True)
df
```

```python
df["Salary"].fillna(0,inplace=True)
df
```

```python
df["Gender"].fillna(0,inplace=True)
df
```

```python
df.dropna(subset=["Education Level"], inplace=True)
df
```

```python
df.dropna(subset=["Years of Experience"], inplace=True)
df
```

```python
df.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
x=df["Years of Experience"].to_frame()
y=df["Salary"].values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_train, y_train, color='brown', label='Trainingpoints') 
plt.scatter(x_test,y_test,color='yellow',label='TestingPoints')
plt.plot(x, model.predict(x), color='red', label='Regression Line') 
plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
plt.title('Salary prediction')
plt.legend()
plt.show()
```
